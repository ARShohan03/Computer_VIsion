```python
import cv2
```

```python
image=cv2.imread("road.jpg")
```

```python
cv2.imshow("original", image)
if(cv2.waitKey()==ord('q')):
	cv2.destroyAllWindows()
```

```python
image_linear=cv2.resize(image, None, fx=5.5, fy=5.5, interpolation= cv2.INTER_LINEAR)
```

```python
image_sized = cv2.resize(image, (300,300))
```

```python
cv2.imshow("Linear", image_linear)
if(cv2.waitKey()==ord('q')):
	cv2.destroyAllWindows()
```

```python
imag_cubic=cv2.resize(image, None, fx=3.3, fy=3.3, interpolation=cv2.INTER_CUBIC)
```

```python
cv2.imshow("Cubic", imag_cubic)
if(cv2.waitKey()==ord('q')):
	cv2.destroyAllWindows()
```
